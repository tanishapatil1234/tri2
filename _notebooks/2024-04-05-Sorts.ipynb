```java
public class Player {
    private String player_name;
    private int player_points;

    public Player(String name, int points) {
        this.player_name = name;
        this.player_points = points;
    }

    // Getters and setters
    public String getPlayerName() {
        return player_name;
    }

    public void setPlayerName(String player_name) {
        this.player_name = player_name;
    }

    public int getPlayerPoints() {
        return player_points;
    }

    public void setPlayerPoints(int player_points) {
        this.player_points = player_points;
    }

    // Insertion sort by points
    public static void insertionSortByPoints(Player[] players) {
        for (int i = 1; i < players.length; i++) {
            Player key = players[i];
            int j = i - 1;
            while (j >= 0 && players[j].getPlayerPoints() > key.getPlayerPoints()) {
                players[j + 1] = players[j];
                j = j - 1;
            }
            players[j + 1] = key;
        }
    }

    // Insertion sort by name
    public static void insertionSortByName(Player[] players) {
        for (int i = 1; i < players.length; i++) {
            Player key = players[i];
            int j = i - 1;
            while (j >= 0 && players[j].getPlayerName().compareTo(key.getPlayerName()) > 0) {
                players[j + 1] = players[j];
                j = j - 1;
            }
            players[j + 1] = key;
        }
    }

    // Selection sort by points
    public static void selectionSortByPoints(Player[] players) {
        for (int i = 0; i < players.length - 1; i++) {
            int minIndex = i;
            for (int j = i + 1; j < players.length; j++) {
                if (players[j].getPlayerPoints() < players[minIndex].getPlayerPoints()) {
                    minIndex = j;
                }
            }
            Player temp = players[minIndex];
            players[minIndex] = players[i];
            players[i] = temp;
        }
    }

    // Selection sort by name
    public static void selectionSortByName(Player[] players) {
        for (int i = 0; i < players.length - 1; i++) {
            int minIndex = i;
            for (int j = i + 1; j < players.length; j++) {
                if (players[j].getPlayerName().compareTo(players[minIndex].getPlayerName()) < 0) {
                    minIndex = j;
                }
            }
            Player temp = players[minIndex];
            players[minIndex] = players[i];
            players[i] = temp;
        }
    }

    // Merge sort by points
    public static void mergeSortByPoints(Player[] players, int left, int right) {
        if (left < right) {
            int middle = (left + right) / 2;
            mergeSortByPoints(players, left, middle);
            mergeSortByPoints(players, middle + 1, right);
            mergeByPoints(players, left, middle, right);
        }
    }

    private static void mergeByPoints(Player[] players, int left, int middle, int right) {
        int n1 = middle - left + 1;
        int n2 = right - middle;

        Player[] L = new Player[n1];
        Player[] R = new Player[n2];

        for (int i = 0; i < n1; ++i)
            L[i] = players[left + i];
        for (int j = 0; j < n2; ++j)
            R[j] = players[middle + 1 + j];

        int i = 0, j = 0;

        int k = left;
        while (i < n1 && j < n2) {
            if (L[i].getPlayerPoints() <= R[j].getPlayerPoints()) {
                players[k] = L[i];
                i++;
            } else {
                players[k] = R[j];
                j++;
            }
            k++;
        }

        while (i < n1) {
            players[k] = L[i];
            i++;
            k++;
        }

        while (j < n2) {
            players[k] = R[j];
            j++;
            k++;
        }
    }

    // Merge sort by name
    public static void mergeSortByName(Player[] players, int left, int right) {
        if (left < right) {
            int middle = (left + right) / 2;
            mergeSortByName(players, left, middle);
            mergeSortByName(players, middle + 1, right);
            mergeByName(players, left, middle, right);
        }
    }

    private static void mergeByName(Player[] players, int left, int middle, int right) {
        int n1 = middle - left + 1;
        int n2 = right - middle;

        Player[] L = new Player[n1];
        Player[] R = new Player[n2];

        for (int i = 0; i < n1; ++i)
            L[i] = players[left + i];
        for (int j = 0; j < n2; ++j)
            R[j] = players[middle + 1 + j];

        int i = 0, j = 0;

        int k = left;
        while (i < n1 && j < n2) {
            if (L[i].getPlayerName().compareTo(R[j].getPlayerName()) <= 0) {
                players[k] = L[i];
                i++;
            } else {
                players[k] = R[j];
                j++;
            }
            k++;
        }

        while (i < n1) {
            players[k] = L[i];
            i++;
            k++;
        }

        while (j < n2) {
            players[k] = R[j];
            j++;
            k++;
        }
    }

    // Quick sort by points
    public static void quickSortByPoints(Player[] players, int low, int high) {
        if (low < high) {
            int pi = partitionByPoints(players, low, high);

            quickSortByPoints(players, low, pi - 1);
            quickSortByPoints(players, pi + 1, high);
        }
    }

    private static int partitionByPoints(Player[] players, int low, int high) {
        int pivot = players[high].getPlayerPoints();
        int i = (low - 1);
        for (int j = low; j < high; j++) {
            if (players[j].getPlayerPoints() < pivot) {
                i++;

                Player temp = players[i];
                players[i] = players[j];
                players[j] = temp;
            }
        }

        Player temp = players[i + 1];
        players[i + 1] = players[high];
        players[high] = temp;

        return i + 1;
    }

    // Quick sort by name
    public static void quickSortByName(Player[] players, int low, int high) {
        if (low < high) {
            int pi = partitionByName(players, low, high);

            quickSortByName(players, low, pi - 1);
            quickSortByName(players, pi + 1, high);
        }
    }

    private static int partitionByName(Player[] players, int low, int high) {
        String pivot = players[high].getPlayerName();
        int i = (low - 1);
        for (int j = low; j < high; j++) {
            if (players[j].getPlayerName().compareTo(pivot) < 0) {
                i++;

                Player temp = players[i];
                players[i] = players[j];
                players[j] = temp;
            }
        }

        Player temp = players[i + 1];
        players[i + 1] = players[high];
        players[high] = temp;

        return i + 1;
    }

    // Bubble sort by points
    public static void bubbleSortByPoints(Player[] players) {
        int n = players.length;
        for (int i = 0; i < n - 1; i++) {
            for (int j = 0; j < n - i - 1; j++) {
                if (players[j].getPlayerPoints() > players[j + 1].getPlayerPoints()) {
                    // Swap players[j] and players[j+1]
                    Player temp = players[j];
                    players[j] = players[j + 1];
                    players[j + 1] = temp;
                }
            }
        }
    }

    // Bubble sort by name
    public static void bubbleSortByName(Player[] players) {
        int n = players.length;
        for (int i = 0; i < n - 1; i++) {
            for (int j = 0; j < n - i - 1; j++) {
                if (players[j].getPlayerName().compareTo(players[j + 1].getPlayerName()) > 0) {
                    // Swap players[j] and players[j+1]
                    Player temp = players[j];
                    players[j] = players[j + 1];
                    players[j + 1] = temp;
                }
            }
        }
    }

    // Main method for testing
    public static void main(String[] args) {
        Player[] players = {
                new Player("Luna", 50),
                new Player("Rachit", 30),
                new Player("Tanisha", 40),
                new Player("Vivian", 20),
                new Player("Zheo", 45)
        };

        System.out.println("Before sorting:");
        for (Player player : players) {
            System.out.println(player.getPlayerName() + " : " + player.getPlayerPoints());
        }

        // Sort by points
        System.out.println("\nSorting by points:");
        Player.insertionSortByPoints(players.clone());
        displayPlayers(players, "Insertion sort by points");

        Player.selectionSortByPoints(players.clone());
        displayPlayers(players, "Selection sort by points");

        Player.mergeSortByPoints(players.clone(), 0, players.length - 1);
        displayPlayers(players, "Merge sort by points");

        Player.quickSortByPoints(players.clone(), 0, players.length - 1);
        displayPlayers(players, "Quick sort by points");

        Player.bubbleSortByPoints(players.clone());
        displayPlayers(players, "Bubble sort by points");

        // Sort by name
        System.out.println("\nSorting by name:");
        Player.insertionSortByName(players.clone());
        displayPlayers(players, "Insertion sort by name");

        Player.selectionSortByName(players.clone());
        displayPlayers(players, "Selection sort by name");

        Player.mergeSortByName(players.clone(), 0, players.length - 1);
        displayPlayers(players, "Merge sort by name");

        Player.quickSortByName(players.clone(), 0, players.length - 1);
        displayPlayers(players, "Quick sort by name");

        Player.bubbleSortByName(players.clone());
        displayPlayers(players, "Bubble sort by name");
    }

    private static void displayPlayers(Player[] players, String sortType) {
        System.out.println("\nAfter " + sortType + ":");
        for (Player player : players) {
            System.out.println(player.getPlayerName() + " : " + player.getPlayerPoints());
        }
    }
}


Player.main(null);
```
